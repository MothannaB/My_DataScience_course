```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
fig1, axes1 = plt.subplots(figsize=(7,5))
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
temperature = [20,21,30,40,30]

axes1.plot(days, temperature, c="red")
axes1.set_title("week temperature report")
axes1.set_xlabel('Day')
axes1.set_ylabel('Temperature')
```

```python
fig2, axes2 = plt.subplots(figsize=(7,5))

scores = np.random.randint(1,7, 6000)
axes2.hist(scores, align="left")
```

```python
fig3, axes3 = plt.subplots(figsize=(7,5))
price_list = {'Red':374, 'Blue':952, 'Yellow':20, 'Green':300}
fav_color= ['Red','Blue', 'Yellow', 'Green']
fav_color_count = [30, 90, 20, 45]
axes3.bar(fav_color, fav_color_count)
axes3.set_title('Class favorite color')
axes3.set_xlabel("Color")
axes3.set_ylabel("Votes")
```
